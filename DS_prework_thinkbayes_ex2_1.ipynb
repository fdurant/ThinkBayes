```python
from thinkbayes import Pmf

class Bowl(object):
    """A bowl that contains a mix of cookies from one or more flavors"""
    
    def __init__(self, name, contents = {}):
        """Initialize self.

        name: string identifier of the bowl
        contents: dictionary with key=flavor and value=number of cookies of that flavor
        """
        
        self.name = name
        self.contents = contents
        
    def takeOneCookie(self, flavor):
        """Take one cookie from the bowl, and update its status

        flavor: string identifier of a flavor
        """

        if flavor not in self.contents.keys():
            raise Exception("%s is not a valid flavor; allowed flavors are [%s]" % (flavor,",".join(self.contents.keys())))
        if (self.contents[flavor] > 1):
            self.contents[flavor] -= 1
        else:
            raise Exception("not enough cookies of flavor %s left to remove one" % flavor)

    def getLikelihood(self, flavor):
        """get the likelihood of drawing a cookie with a specific flavor from the bowl

        flavor: string identifier of a flavor
        """

        totalNrCookies = 0
        for k,v in self.contents.iteritems():
            totalNrCookies += v
        if totalNrCookies == 0:
            raise Exception("no cookies in the bowl, so can't return a likelihood")
        return float(self.contents[flavor])/totalNrCookies
    
    def getName(self):
        """Get the name of the bowl
        """

        return self.name
    
    def __repr__(self):
        description = "%s with:" % self.name
        for k,v in self.contents.iteritems():
            description += " %d %s; " % (v, k)
        return description

class Cookie(Pmf):
    """A map from bowl object to probability."""

    def __init__(self, hypos):
        """Initialize self.

        hypos: sequence of bowl objects
        """
        Pmf.__init__(self)
        for hypo in hypos:
            self.Set(hypo, 1)
        self.Normalize()

    def Update(self, data):
        """Updates the PMF with new data.

        data: string cookie type
        """
        for hypo in self.Values():
            hypo.takeOneCookie(data)
            like = hypo.getLikelihood(data)
            self.Mult(hypo, like)
        print 'took one %s cookie' % data
        self.Normalize()

def main():
    
    # Create two bowls, each with a different mix
    b1 = Bowl('Bowl 1',{'vanilla':30, 'chocolate':10})
    b2 = Bowl('Bowl 2', {'vanilla':20, 'chocolate':20})
    
    # Each bowl is a hypothesis
    hypos = [b1,b2]

    pmf = Cookie(hypos)

    # Show the posterior probabilities, before any cookie is drawn
    for hypo, prob in pmf.Items():
        print hypo, prob

    # Draw a vanilla cookie
    pmf.Update('vanilla')

    # And show the effect on the posterior probabilities
    for hypo, prob in pmf.Items():
        print hypo, prob

    # Now draw a chocolate cookie
    pmf.Update('chocolate')

    # And again show the effect on the posterior probabilities
    for hypo, prob in pmf.Items():
        print hypo, prob

    # Finally, draw a second vanilla cookie
    pmf.Update('vanilla')

    # And, once more, show the effect on the posterior probabilities
    for hypo, prob in pmf.Items():
        print hypo, prob

if __name__ == '__main__':
    main()
```
